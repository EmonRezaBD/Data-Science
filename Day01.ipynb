```
import numpy as np
from matplotlib import pyplot as plt 

ys = 200+ np.random.randn(100) # Generate random values.
print(ys)
x = [x for x in range(len(ys))]

plt.plot(x, ys,'-') #will print 2D Line
plt.fill_between(x, ys, 195, where=(ys>195), facecolor='g', alpha=0.6 ) #will fill the graph

plt.title("Sample Visualization")
plt.show()
```
